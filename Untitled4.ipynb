```python
import gradio as gr
import anthropic

client = anthropic.Anthropic()
score = {"correct": 0, "total": 0}

def build_system_prompt(mode, topic, difficulty):
    if mode == "Assess Me":
        return (f"You are a strict CS professor at {difficulty} difficulty. "
                f"Quiz the user one question at a time on {topic}. "
                f"After each answer, start your reply with CORRECT or INCORRECT, "
                f"explain briefly, then ask the next question.")
    return "You are a helpful, concise assistant for a CS engineering student."

def respond(message, chat_history, mode, topic, difficulty):
    system_prompt = build_system_prompt(mode, topic, difficulty)
    messages = []
    for user_msg, bot_msg in chat_history:
        messages.append({"role": "user", "content": user_msg})
        messages.append({"role": "assistant", "content": bot_msg})
    messages.append({"role": "user", "content": message})

    chat_history.append([message, ""])
    partial = ""
    with client.messages.stream(
        model="claude-sonnet-4-6",
        max_tokens=500,
        system=system_prompt,
        messages=messages,
    ) as stream:
        for text in stream.text_stream:
            partial += text
            chat_history[-1][1] = partial
            yield chat_history, score_text()

    if mode == "Assess Me":
        if partial.strip().upper().startswith("CORRECT"):
            score["correct"] += 1
        score["total"] += 1

def score_text():
    if score["total"] == 0:
        return "Score: вЂ”"
    return f"Score: {score['correct']}/{score['total']}"

def reset_score():
    score["correct"] = 0
    score["total"] = 0
    return [], score_text()

with gr.Blocks(title="Study Assistant Bot") as demo:
    gr.Markdown("## рџЋ“ AI Study Assistant вЂ” Powered by Claude")
    with gr.Row():
        mode = gr.Radio(["Chat", "Assess Me"], value="Chat", label="Mode")
        topic = gr.Textbox(label="Topic (for Assess Me)", placeholder="e.g. Turing Machines")
        difficulty = gr.Dropdown(["Easy", "Medium", "Hard"], value="Medium", label="Difficulty")

    score_display = gr.Textbox(label="Score", value="Score: вЂ”", interactive=False)
    chatbot = gr.Chatbot(height=450)
    msg = gr.Textbox(label="Your message", placeholder="Type here and press Enter")
    clear = gr.Button("Reset Conversation & Score")

    msg.submit(respond, [msg, chatbot, mode, topic, difficulty], [chatbot, score_display])
    msg.submit(lambda: "", None, msg)
    clear.click(reset_score, None, [chatbot, score_display])

demo.launch(share=True, debug=True)
```
